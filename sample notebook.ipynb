```python
import dlt

@dlt.table(
    name="dev.sales.sample_data1",
    comment="Sample data loaded from CSV",
    table_properties={"delta.overwrite": "true"}
)
def sample_data():
    df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load("s3://aws-sample-storage-bucket/sample_data.csv")
    return df
```
